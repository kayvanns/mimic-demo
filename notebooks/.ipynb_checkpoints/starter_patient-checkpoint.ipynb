```python
import pandas as pd
import matplotlib.pyplot as plt
path = "../data/mimic-iv-clinical-database-demo-2.2/hosp/"
path2 = "../data/mimic-iv-clinical-database-demo-2.2/icu/"
```

```python
## reading files and creating dataframes for tables
files = ["admissions", "patients", "labevents", "d_labitems", "prescriptions","pharmacy","transfers", "diagnoses_icd","d_icd_diagnoses"]
dfs = {}

for name in files:
    dfs[name] = pd.read_csv(path + f"{name}.csv.gz")

admissions = dfs["admissions"]
patients = dfs["patients"]
labs = dfs["labevents"]
d_labitems = dfs["d_labitems"]
prescriptions = dfs["prescriptions"]
pharmacy = dfs["pharmacy"]
transfers=dfs["transfers"]
diagnoses = dfs["diagnoses_icd"]
d_diagnoses = dfs["d_icd_diagnoses"]

files2 = ["icustays", "inputevents", "outputevents", "procedureevents","chartevents", "datetimeevents", "d_items"]
dfs2 = {}

for name in files2:
    dfs2[name] = pd.read_csv(path2 + f"{name}.csv.gz")
icustays = dfs2["icustays"]
inputevents = dfs2["inputevents"]
outputevents = dfs2["outputevents"]
procedureevents = dfs2["procedureevents"]
chartevents = dfs2["chartevents"]
datetimeevents = dfs2["datetimeevents"]
d_items = dfs2["d_items"]
```

```python
## getting patient id for first patient
patient  = patients.iloc[[0],: ]
patient_id = int(patient.loc[0,"subject_id"])
patient
```

```python
## getting patient's hospital admissions
patient_admissions = admissions[admissions["subject_id"]==patient_id].copy()
patient_admissions
```

```python
## getting patient's transfers during hospitalization
patient_transfers = transfers[transfers["subject_id"]==patient_id].copy()
patient_transfers
```

```python
## getting patient's lab events and merging to get details
patient_labs = labs[labs["subject_id"]==patient_id].copy()
named_labs = patient_labs.merge(d_labitems,on="itemid")
named_labs.head()
```

```python
## getting patient's icu stay and stay_id
patient_icu = icustays[icustays["subject_id"]==patient_id].copy()
stay_id = int(patient_icu.iloc[0,2])
patient_icu
```

```python
## converting recording time to datetime object, getting patient's icu chartevents and merging to get details
chartevents["charttime"] = pd.to_datetime(chartevents["charttime"])
patient_chartevents = chartevents[chartevents["stay_id"]==stay_id].copy()
named_chartevents = patient_chartevents.merge(d_items,on="itemid")
named_chartevents.head()
```

```python
## playing around with numerics
numeric_chart = named_chartevents[named_chartevents["param_type"]=="Numeric"].copy()
hr = numeric_chart[numeric_chart["label"] == "Heart Rate"]
hr = hr.sort_values("charttime")
fig,ax= plt.subplots()
ax.plot(hr["charttime"], hr["valuenum"])
ax.set_title("Heart Rate Over ICU Stay")
ax.set_ylabel("BPM")
ax.set_xlabel("Time")
ax.tick_params(axis="x", rotation=90)
plt.show()
```

```python
patients_diagnoses = diagnoses[diagnoses["subject_id"].isin(patients_list)].copy()
patients_diagnoses_labelled = patients_diagnoses.merge(d_diagnoses, on=["icd_code","icd_version"])
unique_patient_diagnoses = patients_diagnoses_labelled.drop_duplicates(
    subset=["subject_id", "icd_code"]
)
top_diagnoses = (
    unique_patient_diagnoses["long_title"]
    .value_counts()
    .reset_index().rename(columns={'long_title': 'Diagnoses', 'count': 'Count'})
)
top_diagnoses.to_csv("unique_diagnoses_count.csv")
```
